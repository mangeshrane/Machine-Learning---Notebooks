```python
helper.MONTHS
```

```python
plt.plot(months)
plt.xticks(np.arange(len(months)), helper.MONTHS);
```

```python
plt.plot(months)
plt.xticks([1, 2, 3, 8], helper.MONTHS[[1, 2, 3, 8]]);
```

```python
plt.plot(months)
plt.xticks([low, high], helper.MONTHS[[low, high]]);
```
